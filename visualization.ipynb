```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("train_result.csv")
plt.figure(figsize=(40, 30))
plt.subplot(331)
plt.title("Train Loss")
plt.xlabel("Episodes")
plt.ylabel("Train Loss");
x = range(1, len(data['Train loss'].values) + 1)
#plt.scatter(data.index, data['Train loss'], 'g', label='train loss')
plt.plot(data.index, data['Train loss'].values, '-g', label='train loss');

plt.figure(figsize=(40, 30))
plt.subplot(332)
plt.title("Mean Reward")
plt.xlabel("Episodes")
plt.ylabel("Mean Reward");
plt.plot(data.index, data['Mean Reward'].values, '-b', label='val loss');
plt.legend(loc='best')

plt.figure(figsize=(40, 30))
plt.subplot(333)
plt.title("Total Reward, ")
plt.xlabel("Episodes")
plt.ylabel("Total Reward");
plt.plot(data.index, data['Total Reward'].values, '-b', label='val loss');
plt.legend(loc='best')

plt.show()
```
